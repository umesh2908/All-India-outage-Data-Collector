```python
import pandas as pd
import numpy as pd
import seaborn as sns
import tabula
import io
```

```python
from datetime import datetime, timedelta

# Define the start and end dates in DD.MM.YY format
start_date = "01.05.23"
end_date = "08.05.23"

# Convert the start and end dates into datetime objects
start_date_obj = datetime.strptime(start_date, "%d.%m.%y")
end_date_obj = datetime.strptime(end_date, "%d.%m.%y")

# Create an empty list to store the generated dates
dates = []


current_date = start_date_obj
while current_date <= end_date_obj:
    dates.append(current_date.strftime("%d.%m.%y"))
    current_date += timedelta(days=1)

# Print the list of generated dates
print(dates)
```

# NLDC Daily Reports

```python
base_url = "https://report.grid-india.in/ReportData/Daily%20Report/PSP%20Report/2023-2024/May%202023/"
type=".pdf"

url_list = ["{}{}{}{}".format(base_url, str(dates[i]),"_NLDC_PSP",type) for i in range(0,len(dates))]
s=[]
for url in url_list:
    print (url)
    s.append(url)
```

```python
import os
import requests

# Define the path of the directory where you want to save the downloaded file
folder_name = 'NLDC_Daily_reports'
os.makedirs(folder_name, exist_ok=True)

for url in s:
    # Extract the filename from the URL
    filename = url.split('/')[-1]

    # Join the directory path and the filename to create the full file path
    filepath = os.path.join(folder_name, filename)

     # Check if the file already exists in the folder
    if os.path.exists(filepath):
        print(f'The file {filename} already exists in the {folder_name} directory.')
    else:
        # Send an HTTP GET request to the URL and save the content to a file
        response = requests.get(url)
        with open(filepath, 'wb') as f:
            f.write(response.content)

        print(f'The file {filename} has been downloaded and saved in the {folder_name} directory.')
```

# NRLDC Report 

```python
today = datetime.today()

# Format the date as DD-MM-YYYY
dt = today.strftime("%d-%m-%Y")

# Print the formatted date
print("Today's date is:", dt)
```

```python
import requests
import pandas as pd
import os
from bs4 import BeautifulSoup

url = "https://oms.nrldc.in/outageReport/genUnitReport.php?start={}&dt={}&owner=&eltype=".format(dt,dt)

# create a directory to store the downloaded files
directory = "NRLDC_daily_reports"
if not os.path.exists(directory):
    os.makedirs(directory)

#header=['S.No']
df=pd.read_html(url,header=None)
df
```

```python
Planned=df[0].set_axis(['S.No.','Station','State','Owner','Unit No.','Cap.(MW)','Reasons','Outage','Time','Expected revival'],axis=1,inplace=False)
Forced=df[1].set_axis(['S.No.','Station','State','Owner','Unit No.','Cap.(MW)','Reasons','Outage','Time','Expected revival'],axis=1,inplace=False)
```

```python
Planned.dropna()
Forced.dropna()
```

```python
NRLDC=pd.concat([Planned,Forced])
NRLDC
```

```python
#Save NRLDC data in the NRLDC folder
if not os.path.exists("NRLDC_daily_reports"):
    os.makedirs("NRLDC_daily_reports")

# path to the new directory
path = "NRLDC_daily_reports/"

NRLDC.to_excel(os.path.join(path, "NRLDC_{}.xlsx".format(dt)), index=False)
```
